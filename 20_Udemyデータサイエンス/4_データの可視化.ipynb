# カーネル密度関数(滑らかなヒストグラム)

```python
dataset = randn(25)
sns.kdeplot(dataset)
```

```python
dataset
```

##### バンド幅を変更

```python
sns.rugplot(dataset,color='black')
# バンド幅を変える
for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset,bw=bw,label=bw)
```

##### カーネル関数

```python
kernel_options = ["biw", "cos", "epa", "gau", "tri", "triw"]

for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label=kern)
```

```python
for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label=kern,shade=True,alpha=0.5)
```

##### 軸を入れ替える

```python
sns.kdeplot(dataset,vertical=True)
```

##### 累積分布関数（ヒストグラムの積み上げ）

```python
plt.hist(dataset, cumulative=True)
```

```python
sns.kdeplot(dataset,cumulative=True)
```

##### 二次元の平面

```python
mean = [0,0] # 2次元平面上の中心
cov = [[1,0],[0,100]]# それぞれの分散
# 多変量正規分布
dataset2 = np.random.multivariate_normal(mean,cov,1000)
dframe = pd.DataFrame(dataset2,columns=['X','Y'])
sns.kdeplot(dframe)
```

# ボックスプロット・バイオリンプロット

```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100) + 2 # 平均をずらすため
sns.distplot(data1)
sns.distplot(data2)
```

##### ボックスプロット

```python
sns.boxplot(data=[data1,data2])
```

##### バイオリンプロット

```python
data1 = stats.norm(0,5).rvs(100) # 正規分布
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])# γ分布

sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# バイオリンプロットだと違いがわかる
sns.violinplot(data =[data1,data2])
```

```python
sns.violinplot(data=data2,bw=0.01) # バンド幅を細かく
```

```python
sns.violinplot(data=data1,inner="stick") # rugプロットのように
```

# 回帰とプロット

```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset("tips") #Seabornのサンプルデータ
```

```python
tips.head()
```

```python
sns.lmplot("total_bill","tip",tips) #回帰直線をプロット
```

```python
#4次関数や、パラメータの変更もできる
sns.lmplot("total_bill", "tip", tips, order=4,
           scatter_kws={"marker": "o", "color": "indianred"},
           line_kws={"linewidth": 1, "color": "blue"})
```

```python
sns.lmplot("total_bill", "tip", tips,fit_reg=False) #プロットのみ
```

##### 離散値

```python
tips["tip_pect"]=100*(tips['tip']/tips['total_bill'])
tips.head() #チップの割合を計算
```

```python
sns.lmplot("size", "tip_pect", tips)
```

```python
sns.lmplot("size", "tip_pect", tips, x_jitter=0.2) #見やすく
```

##### 平均値

```python
sns.lmplot("size", "tip_pect", tips, x_estimator=np.mean);
```

##### 傾向

```python
sns.lmplot("total_bill", "tip_pect", tips
           , hue="sex",markers=["x","o"])
```

```python
sns.lmplot("total_bill", "tip_pect", tips, hue="day")
```

##### 曲線を滑らかにする

```python
sns.lmplot("total_bill", "tip_pect", tips
           , lowess=True, line_kws={"color": 'black'});
```

##### 描画エリアの分割

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey =True)

sns.regplot("total_bill","tip_pect",tips,ax=axis1)
sns.violinplot(y='tip_pect', x='size', data=tips.sort_values('size') ,ax=axis2)
```

# ヒートマップとクラスタリング

```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

##### ヒートマップ

```python
df = sns.load_dataset('flights')
df.head()
```

```python
df = df.pivot('month','year','passengers')
df
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df, annot=True, fmt='d') #数字を書き込む
```

```python
sns.heatmap(df, center=df.loc['January',1955]) #色を変える
```

```python
#描画エリアを分割
f, (axis1, axis2) = plt.subplots(2,1) #領域を分ける

yearly_flights = df.sum()

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

year_dframe = pd.concat((years, flights), axis=1)
year_dframe.columns = ['Year', 'Flights']

sns.barplot('Year', y='Flights', data=year_dframe, ax=axis1)
sns.heatmap(df, cmap='Blues', ax=axis2, cbar_kws={'orientation':'horizontal'})
```

##### クラスタリング
近い者同士を寄せて表示

```python
sns.clustermap(df)
```
